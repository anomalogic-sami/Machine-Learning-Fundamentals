```python
import os
import subprocess

# The correct path from your own command prompt output
graphviz_path = r'D:\Programs\Graphviz\bin'

print("--- Running Diagnostic Test ---")
print("Attempting to use this path:", graphviz_path)

try:
    # This command tries to run 'dot -V' just like you did in the command prompt
    # It uses the system's PATH variable to find it.
    result = subprocess.run(
        ['dot', '-V'],
        capture_output=True,
        text=True,
        check=True,
        shell=True,
        env=dict(os.environ, PATH=os.environ['PATH'] + os.pathsep + graphviz_path)
    )
    print("\n✅ SUCCESS! Python can find and run Graphviz.")
    print("Output from dot -V:")
    # The version info is printed to the 'stderr' stream for this command
    print(result.stderr)

except FileNotFoundError:
    print("\n❌ FAILURE: Python still cannot find 'dot.exe'.")
    print("This means the path is still incorrect or inaccessible to Python.")
    print("Please double-check that the folder exists: ", graphviz_path)

except subprocess.CalledProcessError as e:
    print("\n❌ FAILURE: The command ran but returned an error.")
    print("This might indicate a problem with the Graphviz installation itself.")
    print("Error output:", e.stderr)

except Exception as e:
    print("\n❌ FAILURE: An unexpected error occurred.")
    print(e)
```
